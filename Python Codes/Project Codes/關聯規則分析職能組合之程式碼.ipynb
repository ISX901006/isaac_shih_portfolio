```python
from selenium import webdriver
import pandas as pd
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules 
from mlxtend.preprocessing import TransactionEncoder
excel_file_path = 'output_rules.xlsx'

# 配置ChromeDriver的路徑
service = Service(executable_path='./chromedriver.exe')

# 配置Chrome瀏覽器選項
options = webdriver.ChromeOptions()

# 創建Chrome瀏覽器驅動程序
driver = webdriver.Chrome(service=service, options=options)

# 目標網址
url = ('https://www.104.com.tw/jobs/search/?jobsource=2018indexpoc&ro=0')

# 進入網址
driver.get(url)
# 啟動Chrome瀏覽器
job_data = pd.DataFrame(columns=['Job Title', 'Text Job Tag'])
job_titles  =  [
    "行銷企劃",
    "行銷(企劃)主管",
    "產品行銷企劃",
    "產品企劃",
    "遊戲企劃",
    "產品企劃主管",
    "廣告企劃主管",
    "市場調查/市場分析",
    "國內業務",
    "國外業務",
    "金融業務員",
    "國內業務主管",
    "國外業務主管",
    "專案業務主管",
    "產品管理師",
    "專案助理",
    "業務助理",
    "國貿人員",
    "營運管理師/系統整合/ERP專案師",
    "媒體公關/宣傳採買",
    "媒體公關人員/主管",
    "店長/賣場管理人員",
    "船務/報關人員",
    "經營管理主管",
    "主管特別助理",
    "經營管理（高階）主管",
    "專案經理",
    "專案管理主管",
    "其他專案管理師",
    "儲備幹部",
    "財務分析/財務人員",
    "財務分析/財務主管",
    "金融承銷員",
    "金融營業員",
    "金融研究員",
    "金融交易員",
    "股務人員",
    "理賠人員",
    "銀行辦事員",
    "投資經理人",
    "券商後線人員",
    "金融研究人員",
    "金融專業主管",
    "會計師",
    "財務或會計主管",
    "財務或會計高階主管（人員）",
    "查帳/審計人員",
    "記帳/出納/一般會計",
    "成本會計",
    "主辦會計",
    "財務會計助理",
    "稅務人員",
    "稽核人員",
    "理財專員",
    "統計精算人員",
    "核保/保險內勤人員",
    "融資/信用業務人員",
    "融資/信用服務人員",
    "人力資源主管",
    "人力資源人員",
    "風險管理人員",
    "軟體專案管理師",
    "軟體專案主管",
    "軟體工程師",
    "後端工程師",
    "資料庫管理人員",
    "其他資訊專業人員",
    "電玩程式設計師",
    "工程助理",
    "系統分析師",
    "數位轉型專家",
    "數據分析師",
    "資料科學家",
    "商務拓展經理",
    "ESG永續管理師",
    "人工智能專家",
    "客戶成功專員",
    "營收總監",
    "產品負責人",
    "其他客服人員",
    "客服主管",
    "物管/資材",
    "採購人員",
    "採購助理",
    "採購主管",
    "生產技術/製程工程師",
    "工業工程師/生產線規劃",
    "職業安全衛生管理員",
    "品管/品管工程師",
    "品管/檢驗人員",
    "生產管理主管",
    "生管(特別)助理",
    "生管",
    "行政助理",
    "行政人員",
    "行政主管",
    "法務主管",
    "法務人員",
    "秘書",
    "研究助理",
    "顧問",
    "飯店或餐廳主管",
    "文編/校對/文字工作者",
    "公家機關相關人員"
]

all_rules = pd.DataFrame(columns=['Job Title'])
links = []
for job_title in job_titles:
    # 清空DataFrame
    all_job_data = pd.DataFrame(columns=['Job Title', 'Text Job Tag'])
    driver.get(url)
    address_input = driver.find_element(By.XPATH,"/html/body/main/div[2]/div[1]/form/div[1]/div[1]/input")
    address_input.clear()
    address_input.send_keys(job_title)

    # 點擊按鈕
    get_button =  driver.find_element(By.XPATH,"/html/body/main/div[2]/div[1]/form/button")
    get_button.click()
    time.sleep(2)
    print("職業=",job_title)
   #獲取結果果文本
    
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)
    # 使用find_elements_by_class_name方法查找所有具有特定類名
    elements = driver.find_elements(By.CLASS_NAME, "js-job-link")

    count = 0
    max_count = 100
    # 遍歷所有找到的元素，將超鏈接添加到列表中。
    for element in elements:
        link_url = element.get_attribute("href")
        if "jobsource=2018indexpoc" in link_url:
            links.append(link_url)
    print(len(links))
    jobtag_data = []      
    for link in links:
        driver.get(link)
        time.sleep(2)
        jobtag = driver.find_elements(By.XPATH, '//span[@class="d-inline-block"]')
        #jobtag = driver.find_elements(By.XPATH, '//a[@class="tools text-gray-deep-dark d-inline-block"]')
        jobtag_values = [tag.text.replace('、', '') for tag in jobtag]
        if jobtag_values:
            #print(jobtag_values)
            job_df = pd.DataFrame({'Job Title': [job_title], 'Text Job Tag': [jobtag_values]} )
            count += 1
            
            all_job_data = pd.concat([all_job_data,job_df], ignore_index=True)
        if count >= max_count:
            break
    te = TransactionEncoder()
    te_ary = te.fit(all_job_data['Text Job Tag']).transform(all_job_data['Text Job Tag'])

# 將TransactionEncoder的輸出轉換成DataFrame
    dff = pd.DataFrame(te_ary, columns=te.columns_)

# 使用Apriori算法找出頻繁項目及
    frequent_itemsets = apriori(dff, min_support=0.2, use_colnames=True)

    # 使用Association Rules計算信賴度
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
    rules = rules.sort_values(by='confidence', ascending=False)
    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_columns', None)
    # 顯示信賴度
    top_rules = rules.head(5)
    print(top_rules)
    # 將這個職位的結果添加到all_rules
    top_rules['Job Title'] = job_title
    all_rules = pd.concat([all_rules, top_rules], ignore_index=True)
    dff.drop(dff.index, inplace=True)
    jobtag_values.clear()       
    links.clear()
    time.sleep(5)
    driver.back()
    time.sleep(5)

# 輸出Excel檔

all_rules.to_excel(excel_file_path, index=False)
```
