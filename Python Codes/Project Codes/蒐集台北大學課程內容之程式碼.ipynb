```python
# 1. 引入所需套件
import requests
from bs4 import BeautifulSoup
import openpyxl

workbook = openpyxl.Workbook()
path = ‘Course.txt’
f = open(path, ‘r’)
lines = f.readlines()
i = 0
for line in lines:
      links = line
      urls = [links]
      i = i + 1
      for url in urls:
            response = requests.get(url)
            response.raise_for_status()
            html = response.text
            soup = BeautifulSoup(html, ‘html.parser’)
            target_text = ‘教學目標’
            target_table = None
            table = soup.find(‘table’)
            rows = table.find_all(‘tr’)

            for table in soup.find_all(‘table’):
                  if target_text in str(table):
                      target_table = table
                      break
             if target_table:
                 for idx, row in enumerate(target_table.find_all(‘tr)):
                       if target_text in str(row):
                            k = idx

             if len(rows) >= 8:
                                 second_row = rows[k]
                                 first_row = rows[1]
                                 first_sentence1 = first_row.text.split(‘ ‘)[5]
                                 columns = second_row.find_all(‘td’)
                                 for column in columns:
                                       c = column.text
                                       print(i,  ‘.’ ,  first_sentence1, “\n”)
                                       print(c)

              crawler_data = []
              crawler_data.append([first_sentene1, c])
              sheet = workbook.active

              for row in crawler_data:
                           sheet.append(row)

              excel_file_path = ‘Outline.xlsx’

              workbook.save(excel_file_path)
              print(“已輸入”, excel_file_path, “\n”)
f.close()
```
